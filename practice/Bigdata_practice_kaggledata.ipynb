```python
import pandas as pd
import os


fname1 = os.listdir('data')[0]
fname2 = os.listdir('data')[1]

path1 = 'data/'+fname1
path2 = 'data/'+fname2

df1 = pd.read_csv(path1)
df2 = pd.read_csv(path2)

display(df1)
display(df2)
```

```python
import matplotlib.pyplot as plt

# Exploring data
for col in df1.columns:
    if df1[col].dtype == 'float64':
        plt.boxplot(df1[col])
        plt.title(col)
        plt.show()

for col in df1.columns:
    print(col,'has null?:\n',df1[col].isnull().any())
```

```python
import matplotlib.pyplot as plt

# Exploring data
for col in df2.columns:
    #print(df2[col].dtype)
    plt.hist(df2[col])
    plt.title(col)
    plt.show()
    #if df2[col].dtype in ['float64','int64']:
    #    plt.boxplot(df2[col])
    #    plt.title(col)
    #    plt.show()

```

```python
# Check nulls by column
for col in range(len(df2.columns) - 1):
    #print(col)
    #print(df2.iloc[:,col].isna()[df2.iloc[:,col].isna() == True])
    if df2.iloc[:,col].isna().any():
        print('\'' + df2.columns[col] + '\'', 'has null')
    
        df_isnull = df2.iloc[:,col].isna()
        print('Total number of null rows:',len(df_isnull[df_isnull==True]),'/',len(df2.iloc[:,col]))
        print('Null rows:')
        null_rows = df_isnull[df_isnull==True].index.to_list()
        print(null_rows)
        #print('confirm:',df2.columns[col],'\n',df2.iloc[null_rows[0],:])
        
        print()
```

```python
# Check nulls by row

null_r = 0
for row in df2.index:
    if df2.loc[row].isna().any():
        #print(row,'row has null')
        null_r += 1
        #print()
        
print('Number of rows has null :', null_r, '/', len(df2))
print(round((null_r/len(df2))*100,2),'%')
print('Total number of null :', df2.isna().sum().sum())

# 12.37% is pretty high percentage to delete records
```

```python
# 1. Delete all records(241) that has null
df2_dnull = df2.dropna()
print(df2_dnull.corr()['Positive affect'].abs())
print(len(df2_dnull))

from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Encoding Country name - One Hot Encoding by Pandas
#df2_rnull = pd.get_dummies(df2_rnull)
#print(df2_rnull)

# Encoding Country name By One Hot Encoder
oh_encoder = OneHotEncoder()
country_arr = df2_dnull['Country name'].values.reshape(-1,1)
# print(country_arr)
# Delete Country name column
del df2_dnull['Country name']

enc_country_arr = oh_encoder.fit_transform(country_arr).toarray()
lab_country_arr = oh_encoder.get_feature_names(['Country name'])

# Use ex-index to match
df2_country = pd.DataFrame(enc_country_arr, columns=lab_country_arr, index=df2_dnull.index)

# reform to train dataset
x_dnull = pd.concat([df2_dnull, df2_country], axis = 1)
# drop only year column
del x_dnull['year']

y_dnull = x_dnull.loc[:,['Positive affect','Negative affect']]
del x_dnull['Positive affect']
del x_dnull['Negative affect']
```

```python
# 2. Replace all null data w/ mean of each column
df2_rnull = df2
df2_rnull = df2_rnull.fillna(df2_rnull.mean())
print(df2.mean(),'\n')
print(df2.loc[25],'\n')
print(df2_rnull.loc[25],'\n')

#df2_rnull.loc[:,['Country name']] = \
#df2_rnull.loc[:,['Country name']].apply(LabelEncoder().fit_transform)


# One Hot Encoding by Pandas(Easy)
#df2_rnull = pd.get_dummies(df2_rnull)
#print(df2_rnull)
```

```python
# Compare the two process
df2_var = df2.std()
df2_dnull_var = df2_dnull.std()
df2_rnull_var = df2_rnull.std()

df2_compare = pd.concat([df2_var, df2_dnull_var, df2_rnull_var], axis = 1)
df2_compare.columns = ['origin','delete','replace']

display(df2_compare)

dif1 = (df2_var - df2_dnull_var).abs().sum()
dif2 = (df2_var - df2_rnull_var).abs().sum()
print('delete 방식과 기존 데이터의 표준편차 차이 합  : ', dif1)
print('replace 방식과 기존 데이터의 표준편차 차이 합 : ', dif2)

# 0.3030364062985082 vs 0.12911961879189127
# Replacing is less different from the original data
```

```python
# Preprocessing - delete year, Country name columns
X_rnull = df2_rnull.iloc[:,2:9]

# Normalization

from sklearn.preprocessing import MinMaxScaler

df2_r_scaled = MinMaxScaler().fit_transform(X_rnull.iloc[:,1:]) 
df2_r_scaled = pd.DataFrame(df2_r_scaled, columns = X_rnull.columns[1:])
display(df2_r_scaled)


# Standardization

from sklearn.preprocessing import StandardScaler

df2_r_norm = StandardScaler().fit_transform(X_rnull.iloc[:,1:])
df2_r_norm = pd.DataFrame(df2_r_norm, columns = X_rnull.columns[1:])
display(df2_r_norm)
```

```python
# Linear Regression
from sklearn import linear_model

model_rnull = linear_model.LinearRegression()
model_rnull.fit(X=df2_r_scaled, y=df2_rnull['Positive affect'])

help(linear_model.LinearRegression)
```

```python
# Evaluate Model
print('Intercept:', model_rnull.intercept_)
print('Coefficient:', model_rnull.coef_)


print(model_rnull.predict(X=df2_r_scaled.loc[15:18]))
print(df2_rnull['Positive affect'][15:19])
```
