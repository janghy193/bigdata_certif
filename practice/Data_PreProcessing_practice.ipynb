```python
# Check nulls by row

null_r = 0
for row in df2.index:
    if df2.loc[row].isna().any():
        #print(row,'row has null')
        null_r += 1
        #print()
        
print('Number of rows has null :', null_r, '/', len(df2))
print(round((null_r/len(df2))*100,2),'%')
print('Total number of null :', df2.isna().sum().sum())

# 12.37% is pretty high percentage to delete records
```
